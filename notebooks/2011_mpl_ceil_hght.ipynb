```python
%%time
cbh_mpl_11 = []
time_mpl_11 = []
cbh_ceil_11 = []
time_ceil_11 = []

for mpl in mpl_11:
    mpl_ds = act.io.armfiles.read_netcdf(mpl)
    new_mpl = mpl_ds.resample(time='1min').interpolate('linear')
#     time_mpl_11.append(new_mpl.time.copy().shape)
    nmpl_cbh = np.ma.array(new_mpl.cloud_base.values.copy(), dtype=np.float32)
    nmpl_time = np.ma.array(new_mpl.time.values.copy(), dtype=np.datetime64)
    for i in range(len(nmpl_cbh)):
        cbh_mpl_11.append(nmpl_cbh[i])
        time_mpl_11.append(nmpl_time[i])

for ceil in ceil_11:
    ceil_ds = act.io.armfiles.read_netcdf(ceil)
    new_ceil = ceil_ds.resample(time='1min').interpolate('linear')
#     time_ceil_11.append(new_ceil.time.copy().shape)
    nceil_cbh = np.ma.array(new_ceil.first_cbh.values.copy(), dtype=np.float32)
    nceil_time = np.ma.array(new_ceil.time.values.copy(), dtype=np.datetime64)
    for i in range(len(nceil_cbh)):
        cbh_ceil_11.append(nceil_cbh[i])
        time_ceil_11.append(nceil_time[i])
```

```python
mm_11 = np.ma.array(cbh_mpl_11, dtype=np.float32)
cc_11 = np.ma.array(cbh_ceil_11, dtype=np.float32)
tm_11 = np.ma.array(time_mpl_11, dtype=np.datetime64)
tc_11 = np.ma.array(time_ceil_11, dtype=np.datetime64)

mcbh_11 = mm_11[np.ma.where((cc_11 >= 500.) & (cc_11 <= 7620.))[0]]
ccbh_11 = cc_11[np.ma.where((cc_11 >= 500.) & (cc_11 <= 7620.))[0]]
mtime_11 = tm_11[np.ma.where((cc_11 >= 500.) & (cc_11<= 7620.))[0]]
ctime_11 = tc_11[np.ma.where((cc_11 >= 500.) & (cc_11 <= 7620.))[0]]

a_11 = ccbh_11[np.ma.where((mcbh_11 >= .5) & (mcbh_11 <= 7.620))[0]]
b_11 = mcbh_11[np.ma.where((mcbh_11 >= .5) & (mcbh_11 <= 7.620))[0]]
c_11 = ctime_11[np.ma.where((mcbh_11 >= .5) & (mcbh_11 <= 7.620))[0]]
d_11 = mtime_11[np.ma.where((mcbh_11 >= .5) & (mcbh_11 <= 7.620))[0]]
```
